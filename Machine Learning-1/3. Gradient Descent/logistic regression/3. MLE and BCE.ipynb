### Maximum Likelihood Estimation (MLE) and Binary Cross-Entropy (BCE)

Maximum Likelihood Estimation (MLE) and Binary Cross-Entropy (BCE) are closely related in the context of training models, particularly for binary classification tasks. To understand their relationship, let's break down each concept:

#### Maximum Likelihood Estimation (MLE)
- **MLE** is a method for estimating the parameters of a statistical model. Given a dataset, MLE aims to find the parameters that maximize the likelihood of the observed data under the model.
- In simpler terms, MLE chooses the parameters that make the observed data most probable.

#### Binary Cross-Entropy (BCE)
- **BCE** is a loss function commonly used in binary classification tasks, where the goal is to predict one of two possible outcomes (e.g., 0 or 1).
- The BCE loss measures the difference between the predicted probabilities and the actual labels (ground truth). It is defined as:

  $$
  \text{BCE} = -\frac{1}{N} \sum_{i=1}^{N} \left[y_i \cdot \log(p_i) + (1 - y_i) \cdot \log(1 - p_i)\right]
  $$

  where:
  - $y_i$ is the actual label (0 or 1).
  - $p_i$ is the predicted probability of the label being 1.
  - $N$ is the number of samples.

#### Relationship between MLE and BCE

**Link through Likelihood:** In the context of binary classification, if you assume the labels $y_i$ are drawn from a Bernoulli distribution with parameter $p_i$, the likelihood of observing the data is:

$$
L(\theta) = \prod_{i=1}^{N} p_i^{y_i} \cdot (1 - p_i)^{1 - y_i}
$$

where $\theta$ represents the model parameters.

**Log-Likelihood:** To simplify the product, we often work with the log-likelihood:

$$
\log(L(\theta)) = \sum_{i=1}^{N} \left[y_i \cdot \log(p_i) + (1 - y_i) \cdot \log(1 - p_i)\right]
$$

**Negative Log-Likelihood and BCE:** MLE seeks to maximize this log-likelihood. However, in machine learning, we often minimize a loss function instead of maximizing the likelihood. The negative of the log-likelihood becomes the loss function:

$$
\text{Negative Log-Likelihood} = -\log(L(\theta)) = -\sum_{i=1}^{N} \left[y_i \cdot \log(p_i) + (1 - y_i) \cdot \log(1 - p_i)\right]
$$

This expression is precisely the Binary Cross-Entropy (BCE) loss function (up to a scaling factor, which is often $1/N$ for averaging).

#### Summary
- **MLE** aims to find the model parameters that maximize the likelihood of the observed data.
- **BCE** is a specific loss function used in binary classification that can be derived from the negative log-likelihood under the assumption of a Bernoulli distribution.
- Minimizing the BCE loss is equivalent to maximizing the likelihood (using MLE) of the observed binary data under the model.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_flips = 100
n_heads = 80
theta = np.linspace(0, 1, 100)

# Likelihood function
likelihood = (theta**n_heads) * ((1-theta)**(n_flips - n_heads))

# Plot
plt.figure(figsize=(8, 6))
plt.plot(theta, likelihood, label='Likelihood Function')
plt.xlabel('Probability of Heads (θ)')
plt.ylabel('Likelihood')
plt.title('Likelihood Function for Coin Flip')
plt.legend()
plt.show()
```

# Apply log

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_flips = 100
n_heads = 50
theta = np.linspace(0.01, 0.99, 100)  # Avoiding 0 and 1 to prevent log(0) issues

# Likelihood function
likelihood = (theta**n_heads) * ((1-theta)**(n_flips - n_heads))

# Log-Likelihood function
log_likelihood = np.log(likelihood)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(theta, log_likelihood, label='Log-Likelihood Function', color='red')
plt.xlabel('Probability of Heads (θ)')
plt.ylabel('Log-Likelihood')
plt.title('Log-Likelihood Function for Coin Flip')
plt.legend()
plt.grid(True)
```

# BCE in action

```python
# Parameters
y_true = [0, 1]  # True labels
y_pred = np.linspace(0.01, 0.99, 100)  # Predicted probabilities

# BCE calculations for each true label
loss_y_true_0 = -np.log(1 - y_pred)
loss_y_true_1 = -np.log(y_pred)

# Plot
plt.figure(figsize=(8, 6))
plt.plot(y_pred, loss_y_true_0, label='True Label = 0')
plt.plot(y_pred, loss_y_true_1, label='True Label = 1')
plt.xlabel('Predicted Probability (ŷ)')
plt.ylabel('Binary Cross-Entropy Loss')
plt.title('Binary Cross-Entropy Loss Function')
plt.legend()
plt.show()
```
